```python

#get the name of the uploaded file
homeprices = list(uploaded.keys())[0]
# Read the csv file into a pandas Dataframe
homeprices = pd.read_csv(homeprices)
# display the Dataframe to verify it

print (homeprices. head())
```

```python
#Exploratory Data Analysis (EDA):
# Fill in the NaN values
homeprices['bedrooms'].fillna(homeprices['bedrooms'].median(), inplace=True)
homeprices
```

```python
df=homeprices.drop("price", axis="columns")
df
```

```python
price=homeprices.price
price
```

```python
plt.scatter(x=df.area, y= price, marker="+")
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Price of Bedrooms Per Area")
plt.show
```
